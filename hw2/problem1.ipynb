```python
import numpy as np
import matplotlib.pyplot as plt
from itertools import product
import random
```

```python
from epg import fse_signal
```

```python
signal = fse_signal(180, 200, 50, 5, 64)
```

```python
t1st2s = [(200, 50), (200, 100), (1500, 50), (1500, 100), (1500, 1000)]

for t1, t2 in t1st2s:
    plt.plot(-np.imag(fse_signal(180, t1, t2, 5, 64)), label=f'T1={t1}ms, T2={t2}ms')
plt.legend()
plt.xlabel('Echo number')
plt.ylabel('Signal Intensity')
plt.title('$\\alpha=180$°')

plt.show()
```
